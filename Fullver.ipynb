```python
# 기존 문제(2번 목차에 해당)
def knapsack_backtracking(capacity, values, Max_capacity, current_value, current_capacity, index, selected_items):
    global max_value
    global max_selection

    # 기저 사례: 인덱스가 끝에 도달하거나 현재 가치가 최대 가치보다 작을 때
    if index >= len(values) or current_capacity > Max_capacity:
        return

    # 현재 아이템을 선택하는 경우
    if current_capacity + capacity[index] <= Max_capacity:
        current_capacity += capacity[index]
        current_value += values[index]
        selected_items.append(index)

        if current_value > max_value:
            max_value = current_value
            max_selection = selected_items.copy()

        # 현재 아이템을 선택한 후 다음 아이템을 선택하는 경우
        knapsack_backtracking(capacity, values, Max_capacity, current_value, current_capacity, index + 1, selected_items)

        # 현재 아이템을 선택하지 않고 다음 아이템을 선택하는 경우
        current_capacity -= capacity[index]
        current_value -= values[index]
        selected_items.pop()
        knapsack_backtracking(capacity, values, Max_capacity, current_value, current_capacity, index + 1, selected_items)

    # 현재 아이템을 선택하지 않고 다음 아이템을 선택하지 않는 경우
    knapsack_backtracking(capacity, values, Max_capacity, current_value, current_capacity, index + 1, selected_items)

# 학과/학부 이름
items_names = ['건축사회환경공학부', '건축학과', '산업경영공학부', '신소재공학부', '전기전자공학부', 
                    '기계공학부', '화공생명공학과', '융합에너지공학과', '반도체공학과','차세대통신학과']

# 학과별 가치와 인원수
values = [80, 60, 100, 100, 200, 110, 90, 50, 40, 60]
capacity = [100, 40, 50, 120, 250, 130, 100, 30, 30, 30]

# 술집 수용 인원
Max_capacity = 300

# 최대 가치 초기화
max_value = 0
max_selection = []

# 백트래킹 알고리즘 실행
knapsack_backtracking(capacity, values, Max_capacity, 0, 0, 0, [])

# 결과 출력
print("최대 가치:", max_value)
print("선택된 학과들:")
for index in max_selection:
    print(f"- {items_names[index]}")

import time

def calculate_execution_time(func, *args, **kwargs):
    start_time = time.time()
    func(*args, **kwargs)
    end_time = time.time()
    execution_time = end_time - start_time
    return execution_time

# 코드 실행 시간 측정
execution_time = calculate_execution_time(knapsack_backtracking, capacity, values, Max_capacity, 0, 0, 0, [])

# 결과 출력
print("코드 실행 시간:", execution_time, "초")
```

```python
# Greedy_기본
def knapsack_greedy(items, Max_Capacity):
    value_per_weight = [(item['Value'] / item['Capacity'], item) for item in items]
    value_per_weight.sort(key=lambda x: x[0], reverse=True)

    total_value = 0
    total_capacity = 0
    selected_items = []

    for value_ratio, item in value_per_weight:
        if total_capacity + item['Capacity'] <= Max_Capacity:
            total_value += item['Value']
            total_capacity += item['Capacity']
            selected_items.append(item)

    return total_value, total_capacity, selected_items


# 물건 정보
items = [
    {'Major': '건축사회환경공학부', 'Capacity': 100, 'Value': 80},
    {'Major': '건축학과', 'Capacity': 40, 'Value': 60},
    {'Major': '산업경영공학부', 'Capacity': 50, 'Value': 100},
    {'Major': '신소재공학부', 'Capacity': 120, 'Value': 100},
    {'Major': '전기전자공학부', 'Capacity': 250, 'Value': 200},
    {'Major': '기계공학부', 'Capacity': 130, 'Value': 110},
    {'Major': '화공생명공학과', 'Capacity': 100, 'Value': 90},
    {'Major': '융합에너지공학과', 'Capacity': 30, 'Value': 50},
    {'Major': '반도체공학과', 'Capacity': 30, 'Value': 40},
    {'Major': '차세대통신학과', 'Capacity': 30, 'Value': 60}
]

# 배낭의 용량
Max_Capacity = 300

max_value, total_capacity, selected_items = knapsack_greedy(items, Max_Capacity)

print(f"식당에 담을 수 있는 최대 가치: {max_value}")
print(f"식당의 총 용량: {total_capacity}")
print(f"식당의 최대 용량: {Max_Capacity}")
print("선택된 학부들:")
for item in selected_items:
    print(f"전공: {item['Major']}, 인원 수: {item['Capacity']}, 가치: {item['Value']}")
```

### Multiple Knapsack

```python
def Multiple_Knapsack():
    items = [
        {'Major': '건축사회환경공학부', 'Capacity': 100, 'Value': 80},
        {'Major': '건축학과', 'Capacity': 40, 'Value': 60},
        {'Major': '산업경영공학부', 'Capacity': 50, 'Value': 100},
        {'Major': '신소재공학부', 'Capacity': 120, 'Value': 100},
        {'Major': '전기전자공학부', 'Capacity': 250, 'Value': 230},
        {'Major': '기계공학부', 'Capacity': 130, 'Value': 110},
        {'Major': '화공생명공학과', 'Capacity': 100, 'Value': 90},
        {'Major': '융합에너지공학과', 'Capacity': 30, 'Value': 50},
        {'Major': '반도체공학과', 'Capacity': 30, 'Value': 40},
        {'Major': '차세대통신학과', 'Capacity': 30, 'Value': 60}
    ]

    n = len(items)

    capacity = 300

    # 각 물건의 가치와 무게 정보를 사용하여 테이블 초기화
    table = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            item = items[i - 1]
            if item['Capacity'] <= w:
                # 물건을 배낭에 넣을 수 있는 경우, 최대 가치를 선택
                table[i][w] = max(table[i - 1][w], item['Value'] + table[i - 1][w - item['Capacity']])
            else:
                # 물건을 배낭에 넣을 수 없는 경우, 이전 테이블의 값 사용
                table[i][w] = table[i - 1][w]

    # 최대 가치를 찾기 위한 변수 초기화
    max_value = table[n][capacity]
    bag1 = []
    bag2 = []

    # 두 개의 배낭에 어떤 물건들이 들어갔는지 추적
    w1 = w2 = capacity
    for i in range(n, 0, -1):
        if table[i][w1] != table[i - 1][w1]:
            item = items[i - 1]
            bag1.append(item['Major'])
            w1 -= item['Capacity']
        elif table[i][w2] != table[i - 1][w2]:
            item = items[i - 1]
            bag2.append(item['Major'])
            w2 -= item['Capacity']

    # 배낭의 capacity가 얼마나 채워졌는지 계산
    filled_capacity_bag1 = capacity - w1
    filled_capacity_bag2 = capacity - w2


    # 총 배낭 가치 계산
    total_value = sum(item['Value'] for item in items if item['Major'] in bag1 or item['Major'] in bag2)

    # 결과 출력
    print("춘자_지상 배정 학부 목록:", bag1)
    print("춘자_지하 배정 학부 목록:", bag2)
    print("총 가치:", total_value)
    print("춘자_지상의 채워진 정도:", filled_capacity_bag1)
    print("춘자_지하의 채워진 정도:", filled_capacity_bag2)

Multiple_Knapsack()
```

### Add Variable

```python
def knapsack_with_capacity_and_volume(items, capacity, volume_limit):
    n = len(items)
    weights = [item['Capacity'] for item in items]
    values = [item['Value'] for item in items]
    volumes = [item['Orders'] for item in items]

    dp = [[[0] * (volume_limit + 1) for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        for j in range(capacity + 1):
            for k in range(volume_limit + 1):
                if i == 0 or j == 0 or k == 0:
                    dp[i][j][k] = 0
                elif weights[i - 1] <= j and volumes[i - 1] <= k:
                    dp[i][j][k] = max(
                        values[i - 1] + dp[i - 1][j - weights[i - 1]][k - volumes[i - 1]],
                        dp[i - 1][j][k]
                    )
                else:
                    dp[i][j][k] = dp[i - 1][j][k]

    selected_items = []
    selected_values = []
    selected_orders = []
    used_capacity = 0
    used_orders = 0
    i, j, k = n, capacity, volume_limit
    while i > 0 and j > 0 and k > 0:
        if dp[i][j][k] != dp[i - 1][j][k]:
            selected_items.append(items[i - 1]['Major'])
            selected_values.append(items[i - 1]['Value'])
            selected_orders.append(items[i - 1]['Orders'])
            used_capacity += weights[i - 1]
            used_orders += volumes[i - 1]
            j -= weights[i - 1]
            k -= volumes[i - 1]
        i -= 1

    total_value = dp[n][capacity][volume_limit]

    print("식당에 담을 수 있는 최대 가치:", total_value)
    print("식당의 최대 용량:", capacity)
    print("최대 주문량:", volume_limit)
    print("사용된 용량:", used_capacity)
    print("사용된 주문량:", used_orders)
    print("선택된 학부들:")
    for item in selected_items:
        print(item)


# 예시 입력
items = [
    {'Major': '건축사회환경공학부', 'Capacity': 100, 'Value': 80, 'Orders': 90},
    {'Major': '건축학과', 'Capacity': 40, 'Value': 60, 'Orders': 50},
    {'Major': '산업경영공학부', 'Capacity': 50, 'Value': 100, 'Orders': 70},
    {'Major': '신소재공학부', 'Capacity': 120, 'Value': 100, 'Orders': 110},
    {'Major': '전기전자공학부', 'Capacity': 250, 'Value': 230, 'Orders': 200},
    {'Major': '기계공학부', 'Capacity': 130, 'Value': 110, 'Orders': 150},
    {'Major': '화공생명공학과', 'Capacity': 100, 'Value': 90, 'Orders': 70},
    {'Major': '융합에너지공학과', 'Capacity': 30, 'Value': 50, 'Orders': 40},
    {'Major': '반도체공학과', 'Capacity': 30, 'Value': 40, 'Orders': 40},
    {'Major': '차세대통신학과', 'Capacity': 30, 'Value': 60, 'Orders': 50}
]

capacity = 300
volume_limit = 250

knapsack_with_capacity_and_volume(items, capacity, volume_limit)
```
